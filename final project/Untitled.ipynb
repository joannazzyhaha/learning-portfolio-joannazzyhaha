```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function T(t)
def T(t, Ts, T0, k):
    return Ts + (T0 - Ts) * np.exp(-k * t)

Ts = 68  # Initial temperature
T0 = 94.4  # Final temperature
k = 0.16452  # Rate constant
time_values = np.linspace(-8, 8, 300)  # Time values

# Calculate temperature values using the function
temperature_values = T(time_values, Ts, T0, k)


plt.plot(time_values, temperature_values, label='T(t)')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature vs Time')
plt.legend()
plt.grid(True)
plt.show()
```
